```python
#1.1

import math

class Trigonometry:
    @staticmethod
    def cosine(x):
        return math.cos(x)

    @staticmethod
    def sine(x):
        return math.sin(x)

    @staticmethod
    def tangent(x):
        return math.tan(x)

    @staticmethod
    def arcsine(x):
        return math.asin(x)

    @staticmethod
    def arccosine(x):
        return math.acos(x)

    @staticmethod
    def arctangent(x):
        return math.atan(x)

    @staticmethod
    def degrees_to_radians(degrees):
        return math.radians(degrees)

value = float(input("Введите значение в градусах: "))

print("Cosine:", Trigonometry.cosine(value))
print("Sine:", Trigonometry.sine(value))
print("Tangent:", Trigonometry.tangent(value))
print("Arcsine:", Trigonometry.arcsine(0.7071))
print("Arccosine:", Trigonometry.arccosine(0.7071))
print("Arctangent:", Trigonometry.arctangent(1))
print("Radians:", Trigonometry.degrees_to_radians(value))
```

```python
#1.2.2

class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def PrintTree(self):
        if self.left:
            self.left.PrintTree()
        print(self.data)
        if self.right:
            self.right.PrintTree()

    def AddElement(self, Element):
        if self.data:
            if Element < self.data:
                if self.left is None:
                    self.left = Tree(Element)
                else:
                    self.left.AddElement(Element)
            elif Element > self.data:
                if self.right is None:
                    self.right = Tree(Element)
                else:
                    self.right.AddElement(Element)
        else:
            self.data = Element

root = Tree(4)

root.AddElement(2)
root.AddElement(1)
root.AddElement(3)
root.AddElement(6)
root.AddElement(5)
root.AddElement(7)

root.PrintTree()
```
