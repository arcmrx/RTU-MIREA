```python
#1.3.1
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def euclidean_distance(v1, v2):
  return sum((x - y) ** 2 for x, y in zip(v1, v2)) ** 0.5

def sqr_euclidean_distance(v1, v2):
  return sum((x - y) ** 2 for x, y in zip(v1, v2))

def weighted_euclidean_distance(v1, v2, w):
  return sum((x - y) ** 2 * s for x, y, s in zip(v1, v2, w)) ** 0.5

def manhattan_distance(v1, v2):
  return sum(abs(x - y) for (x, y) in zip(v1, v2))

def chebyshev_distance(v1, v2):
  return max(abs(x - y) for (x, y) in zip(v1, v2))

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x1 = np.array([0, 0, 0])
x2 = np.array([3, 3, 3])
x3 = np.array([3, 0, 2])
x4 = np.array([1, 2, 3])
w = np.array([0, 0, 1])

print('Евклидово расстояние(с, о): \t\t', euclidean_distance(x1, x2))
print('Квадрат Евклидова расстояния(с, о): \t', sqr_euclidean_distance(x1, x2))
print('Взвешенное Евклидово расстояние(с, о): \t', weighted_euclidean_distance(x1, x2, w))
print('Хеммингово расстояние(с, о): \t\t', manhattan_distance(x1, x2))
print('Расстояние Чебышева(с, о): \t\t', chebyshev_distance(x1, x2))
print('\n')
print('Евклидово расстояние(с, з): \t\t', euclidean_distance(x1, x3))
print('Квадрат Евклидова расстояния(с, з): \t', sqr_euclidean_distance(x1, x3))
print('Взвешенное Евклидово расстояние(с, з): \t', weighted_euclidean_distance(x1, x3, w))
print('Хеммингово расстояние(с, з): \t\t', manhattan_distance(x1, x3))
print('Расстояние Чебышева(с, з): \t\t', chebyshev_distance(x1, x3))
print('\n')
print('Евклидово расстояние(с, к): \t\t', euclidean_distance(x1, x4))
print('Квадрат Евклидова расстояния(с, к): \t', sqr_euclidean_distance(x1, x4))
print('Взвешенное Евклидово расстояние(с, к): \t', weighted_euclidean_distance(x1, x4, w))
print('Хеммингово расстояние(с, к): \t\t', manhattan_distance(x1, x4))
print('Расстояние Чебышева(с, к): \t\t', chebyshev_distance(x1, x4))
print('\n')
print('Евклидово расстояние(о, з): \t\t', euclidean_distance(x2, x3))
print('Квадрат Евклидова расстояния(о, з): \t', sqr_euclidean_distance(x2, x3))
print('Взвешенное Евклидово расстояние(о, з): \t', weighted_euclidean_distance(x2, x3, w))
print('Хеммингово расстояние(о, з): \t\t', manhattan_distance(x2, x3))
print('Расстояние Чебышева(о, з): \t\t', chebyshev_distance(x2, x3))
print('\n')
print('Евклидово расстояние(о, к): \t\t', euclidean_distance(x2, x4))
print('Квадрат Евклидова расстояния(о, к): \t', sqr_euclidean_distance(x2, x4))
print('Взвешенное Евклидово расстояние(о, к): \t', weighted_euclidean_distance(x2, x4, w))
print('Хеммингово расстояние(о, к): \t\t', manhattan_distance(x2, x4))
print('Расстояние Чебышева(о, к): \t\t', chebyshev_distance(x2, x4))
print('\n')
print('Евклидово расстояние(з, к): \t\t', euclidean_distance(x3, x4))
print('Квадрат Евклидова расстояния(з, к): \t', sqr_euclidean_distance(x3, x4))
print('Взвешенное Евклидово расстояние(з, к): \t', weighted_euclidean_distance(x3, x4, w))
print('Хеммингово расстояние(з, к): \t\t', manhattan_distance(x3, x4))
print('Расстояние Чебышева(з, к): \t\t', chebyshev_distance(x3, x4))
print('\n')

ax.scatter(0, 0, 0, label = 'C') #c
ax.scatter(3, 3, 3, label = 'О') #о
ax.scatter(3, 0, 2, label = 'З') #з
ax.scatter(1, 2, 3, label = 'К') #к
ax.legend()
plt.show()
```

```python
#2.3.1
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
import seaborn as sns

iris = sns.load_dataset('iris')

X_train, X_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size=0.15
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape


def init_model(k, X_train, y_train, X_test):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    print(model)
    y_pred = model.predict(X_test)
    return y_pred


def graph(y_test, X_test, y_pred):
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x='petal_width', y='petal_length', data=iris, hue='species', s=70)
    plt.xlabel('Длина лепестка, см')
    plt.ylabel('Ширина лепестка, см')
    plt.legend(loc=2)
    plt.grid()

    for i in range(len(y_test)):
        if np.array(y_test)[i] != y_pred[i]:
            plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color='red', s=150)


y_pred = init_model(1, X_train, y_train, X_test)
graph(y_test, X_test, y_pred)
y_pred = init_model(5, X_train, y_train, X_test)
graph(y_test, X_test, y_pred)
y_pred = init_model(10, X_train, y_train, X_test)
graph(y_test, X_test, y_pred)
```
