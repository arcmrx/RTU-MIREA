```python
#4.3.2
import math
import numpy as np
import matplotlib.pyplot as plt
f = [(math.sqrt(1 + math.e ** math.sqrt(x) + math.cos(x ** 2))) / abs(1 - (math.sin(x)) ** 3) for x in range(1, 11)]
x = np.arange(0.0, max(f), max(f)/10)
newf = f[:5]
plt.grid()
plt.plot(x, f)
plt.show()
plt.grid()
plt.scatter(x[:5], newf)
plt.show()
```

```python
#4.3.3
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz
f = [(abs(math.cos(x * (math.e ** (math.cos(x) + math.log(x + 1)))))) for x in range(0, 11)]
x = np.arange(0.0, max(f), max(f)/11)
plt.grid()
plt.plot(x, f, c = "r")
plt.fill_between(x, f)
area = trapz(f, x)
print(area)
```

```python
#4.3.5
import math

print("Введите x: \n")
x = float(input())
print("Введите y: \n")
y = float(input())
print("Введите действие:\n",
      "'+' (x + y)\n",
      "'-' (x - y)\n",
      "'*' (x * y)\n",
      "'/' (x / y)\n",
      "'e' (e ^ (x + y))\n",
      "'sin' (sin(x + y))\n",
      "'cos' (cos(x + y))\n",
      "'^' (x ^ y)\n",)
d = str(input())
if d == "+":
  print("x + y = ", x + y)
elif d == "-":
  print("x - y = ", x - y)
elif d == "*":
  print("x * y = ", x * y)
elif d == "/":
  if y == 0:
    print("На ноль делить нельзя!")
  else:
    print("x / y = ", x / y)
elif d == "e":
  print("e ^ (x + y) = ", math.e ** (x + y))
elif d == "sin":
  print("sin(x + y) = ", math.sin(x + y))
elif d == "cos":
  print("cos(x + y) = ", math.cos(x + y))
elif d == "^":
  print("x ^ y = ", x**y)
else:
  print("Ошибка!!!")
```
